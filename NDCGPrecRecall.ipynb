```python
from collections import defaultdict
import math
# Function to read data from a file and create a dictionary
def read_qrel_relevance(file_path):
    qrel_relevance_dict = {}
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split()
            if len(parts) >= 4:
                qid, doc_id, relevance = parts[0], (parts[2]), int(parts[3])
                if qid not in qrel_relevance_dict:
                    qrel_relevance_dict[qid] = {}
                qrel_relevance_dict[qid][doc_id] = relevance
    return qrel_relevance_dict
```

```python
qrels_file_path = './qrel_file/ordered-qrel-msmarco-2019.txt'
scores_file_path = './relevance_file/DAATBM25withRelevance.txt'
```

```python
# Step 1: Create data structures
qrels_data = read_qrel_relevance(qrels_file_path)
scores_data = read_qrel_relevance(scores_file_path)
```

```python
precisions=[]
recalls=[]
common_queries = set(qrels_data.keys()) & set(scores_data.keys())
def intersection_and_length(dict, key, other_list):
    dict_list = dict.get(key, [])
    if not dict_list:
        print(key+"empty")
        display(key+"empty")
    intersection = set(dict_list) & set(other_list)
    return len(intersection)
for qid in common_queries:
    doc_ids=scores_data[qid]
    len_of_intersection=intersection_and_length(qrels_data,qid,doc_ids)
    precisions.append([qid,(len_of_intersection/len(doc_ids))])
    base_recall=len(qrels_data[qid])
    recalls.append([qid,(len_of_intersection/base_recall)if base_recall!=0 else 0])
```

```python
values=[value for _,value in precisions]
mean=sum(values)/len(values)
mean
```
